```python
import pandas as pd
```

```python
nba_data = pd.read_csv('data/out.csv')

X = nba_data.drop(['PTS'], axis=1)
y = nba_data['PTS']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=19)
X_train
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr = RandomForestRegressor(random_state=13, n_jobs=-1, verbose=True)

rfr.fit(X_train, y_train)
```

```python
y_pred = rfr.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, r2_score, mean_squared_error
```

```python
float(mean_absolute_error(y_pred, y_test))
```

```python
float(mean_squared_error(y_pred, y_test))
```

```python
float(r2_score(y_pred, y_test))
```

```python
params = {
    'n_estimators' : [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split' : [2, 5, 10],
    'min_samples_leaf' : [1, 2, 4]
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
rfr_cv = GridSearchCV(estimator=rfr, param_grid=params, scoring="neg_mean_squared_error", n_jobs=-1, verbose=True)
```

```python
rfr_cv.fit(X_train, y_train)
```
